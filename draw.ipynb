Đây là code để vẽ một số biểu đồ sau khi chạy.
Do tài nguyên không đủ và thời gian chạy lâu nên chúng em chỉ thử nghiệm số User Equipment UE = 5 cho đến UE = 15. Nếu UE max khác 15 thì có thể thay đổi biến UE dưới đây.

```python
UE = 15
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

OFFLOADED_TASK

```python
offload_tasks_dqn=[]
offload_tasks_doubledqn=[]
offload_tasks_onl=[]
offload_tasks_off=[]
offload_tasks_semi=[]
offload_tasks_dueldqn=[]
```

```python
for UN in range (5, UE + 1):
    double=pd.read_csv('dataDouble\JOB_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    dqn=pd.read_csv('dataDQN\JOB_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    duel=pd.read_csv('dataDueling\JOB_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    onl=pd.read_csv('dataDQN\JOB_online2_'+str(UN)+'_rho_4.data',sep='\t')
    off=pd.read_csv('dataDQN\JOB_offline2_'+str(UN)+'_rho_4.data',sep='\t')
    semi=pd.read_csv('dataDQN\JOB_semi2_'+str(UN)+'_rho_4.data',sep='\t')
    offload_tasks_dueldqn.append(len(duel))
    offload_tasks_doubledqn.append(len(double))
    offload_tasks_dqn.append(len(dqn))
    offload_tasks_onl.append(len(onl))
    offload_tasks_off.append(len(off))
    offload_tasks_semi.append(len(semi))
```

```python
offload_tasks_doubledqn
```

```python
plt.plot(UE,offload_tasks_onl,label='Online',marker='x')
plt.plot(UE,offload_tasks_off,label='Offline',marker='o')
plt.plot(UE,offload_tasks_semi,label='Semi',marker='*')
plt.plot(UE,offload_tasks_dqn,label='DQN',marker='d')
plt.plot(UE,offload_tasks_doubledqn,label='DoubleDQN',marker='s')
plt.plot(UE,offload_tasks_dueldqn,label='DuelingDQN',marker='v')
plt.xlabel('Number of UE')
plt.ylabel('Offloaded tasks in total')

plt.legend()
plt.show()
```

COMMUNICATION ENERGY

```python
commEnergy_dqn=[]
commEnergy_doubledqn=[]
commEnergy_onl=[]
commEnergy_off=[]
commEnergy_semi=[]
commEnergy_dueldqn=[]
```

```python
for UN in range (5, UE + 1):
    double=pd.read_csv('dataDouble\\USER_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    dqn=pd.read_csv('dataDQN\\USER_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    onl=pd.read_csv('dataDQN\\USER_online2_'+str(UN)+'_rho_4.data',sep='\t')
    off=pd.read_csv('dataDQN\\USER_offline2_'+str(UN)+'_rho_4.data',sep='\t')
    semi=pd.read_csv('dataDQN\\USER_semi2_'+str(UN)+'_rho_4.data',sep='\t')
    duel=pd.read_csv('dataDueling\\USER_DQN2_'+str(UN)+'_rho_4.data',sep='\t')
    commEnergy_doubledqn.append(double[' commEnergy '][len(double)-1])
    commEnergy_dqn.append(dqn[' commEnergy '][len(dqn)-1])
    commEnergy_onl.append(onl[' commEnergy '][len(onl)-1])
    commEnergy_off.append(off[' commEnergy '][len(off)-1])
    commEnergy_semi.append(semi[' commEnergy '][len(semi)-1])
    commEnergy_dueldqn.append(duel[' commEnergy '][len(duel)-1])
```

```python
plt.plot(UE,commEnergy_onl,label='Online',marker='x')
plt.plot(UE,commEnergy_off,label='Offline',marker='o')
plt.plot(UE,commEnergy_semi,label='Semi',marker='*')
plt.plot(UE,commEnergy_dqn,label='DQN',marker='d')
plt.plot(UE,commEnergy_doubledqn,label='DoubleDQN',marker='s')
plt.plot(UE,commEnergy_dueldqn,label='DuelingDQN',marker='v')
plt.xlabel('Number of UE')
plt.ylabel('Communication energy consumption per UE')

plt.legend()
plt.show()
```
